Extracting the duration in seconds for better analysis

```python
import pandas as pd

# Read the original CSV file
df = pd.read_csv("Original.csv")

# Convert 'durationText' (hh:mm:ss) back to total seconds
df['duration'] = pd.to_timedelta(df['durationText']).dt.total_seconds().astype(int)

df.to_csv("Original.csv", index=False)

# Display first few rows
df.head()
```

Changing the data of published ago into days

```python
for index, value in df['publishedText'].items():  
    row = value.split(" ")  
    if row[-2] in ['hours', 'hour']:  
        df.at[index, 'publishedText'] = round(int(row[-3])/24, 2)  
    elif row[-2] in ['months', 'month']:  
        df.at[index, 'publishedText'] = round(int(row[-3])*30, 2)  
    elif row[-2] in ['weeks', 'week']:  
        df.at[index, 'publishedText'] = round(int(row[-3])*7, 2)  
    elif row[-2] in['day','days']:
        df.at[index,'publishedText']=int(row[-3])
df['publishedText'] = pd.to_numeric(df['publishedText'])
```

Checking for outliers in views and published text

```python
import matplotlib.pyplot as plt
import numpy as np


x=df["views"]
x=np.array(x)
y=df["publishedText"]
y=np.array(y)


#Scatter plot (before outlier removal)
plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.5)
plt.xlabel("Views")
plt.ylabel("Published Date (Unix Time)")
plt.title("Scatter Plot of Views vs Published Date")
plt.show()

# Create box plots to detect outliers
fig, ax = plt.subplots()
box = ax.boxplot(x)
plt.title("Boxplot for Views")

# Get outlier values for views
outliers = np.unique(box['fliers'][0].get_ydata())

print("Outliers in Views:", outliers)

# Remove outliers
mask = ~np.isin(x, outliers)  # Mask to keep only non-outlier values
x_cleaned = x[mask]
y_cleaned = y[mask]

# Scatter plot after outlier removal
plt.figure(figsize=(8, 5))
plt.scatter(x_cleaned, y_cleaned, alpha=0.5, color="red")
plt.xlabel("Views")
plt.ylabel("Published Date (Unix Time)")
plt.title("Scatter Plot (Outliers Removed)")
plt.show()
```

Plot between view and publication time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings('ignore')
# Load the dataset (replace 'your_file.csv' with your actual CSV file)
df = pd.read_csv("Original.csv")


# Function to categorize publication time from 'publishedText'
def categorize_time(published_text):
    # Extract the first number found in the text
    match = re.search(r"\d+", published_text)
    
    if match:  # If a number is found, process it
        number = int(match.group())

        if "day" in published_text or "days" in published_text:
            if number <= 7:
                return "1 Week"
            elif number <= 14:
                return "2 Weeks"
            elif number <= 21:
                return "3 Weeks"
            elif number <= 28:
                return "4 Weeks"
        elif "week" in published_text or "weeks" in published_text:
            if number == 1:
                return "1 Week"
            elif number == 2:
                return "2 Weeks"
            elif number == 3:
                return "3 Weeks"
            elif number == 4:
                return "4 Weeks"
    
    return "4+ Weeks"  # Default category for non-matching values

# Apply the categorization function
df["category"] = df["publishedText"].apply(categorize_time)

# Group by category and sum views
df_grouped = df.groupby("category")["views"].sum().reset_index()

# Sort categories for better visualization
order = ["1 Week", "2 Weeks", "3 Weeks", "4 Weeks", "4+ Weeks"]
df_grouped = df_grouped.set_index("category").reindex(order).reset_index()

# Set plot style
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x="category", y="views", data=df_grouped, palette="Reds", legend = False)

# Labels and title
plt.xlabel("Publication Time")
plt.ylabel("Total Views")
plt.title("Bar Graph: Views vs Publication Time (Using publishedText)")

# Show the plot
plt.show()
```

```python
#during the initial phase of publication of video youtube pushes the video through its algorithm to other users
#but with time decraeses the appearance of the video
```
